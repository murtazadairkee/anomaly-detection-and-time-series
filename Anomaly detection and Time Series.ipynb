### Question 1 : What is Dimensionality Reduction? Why is it important in machine learning?
**Answer:**
Dimensionality reduction is the process of reducing the number of input variables or features in a dataset while retaining as much relevant information as possible. It helps simplify models, reduce overfitting, improve computation speed, and make data visualization easier.

It is important because high-dimensional data can lead to the 'curse of dimensionality', where models become complex and inefficient.

### Question 2: Name and briefly describe three common dimensionality reduction techniques.
**Answer:**
1. **Principal Component Analysis (PCA):** Converts correlated variables into a smaller number of uncorrelated variables called principal components.
2. **Linear Discriminant Analysis (LDA):** A supervised technique that reduces dimensions while preserving class separability.
3. **t-Distributed Stochastic Neighbor Embedding (t-SNE):** Non-linear technique that helps visualize high-dimensional data in 2D or 3D.

### Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.
**Answer:**
Clustering is an unsupervised learning technique used to group similar data points into clusters without predefined labels.

Three popular clustering algorithms are:
1. K-Means Clustering
2. Hierarchical Clustering
3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

### Question 4: Explain the concept of anomaly detection and its significance.
**Answer:**
Anomaly detection is the process of identifying data points, events, or observations that deviate significantly from the majority of the data. These anomalies often indicate critical incidents such as fraud, system failures, or security breaches.

It is significant because it helps detect unusual behavior early, preventing potential risks or losses.

### Question 5: List and briefly describe three types of anomaly detection techniques.
**Answer:**
1. **Statistical Methods:** Identify anomalies based on probability distributions (e.g., z-scores, Gaussian models).
2. **Machine Learning-Based Methods:** Use algorithms like Isolation Forest, One-Class SVM.
3. **Distance-Based Methods:** Detect anomalies based on distances between data points, such as k-NN-based detection.

### Question 6: What is time series analysis? Mention two key components of time series data.
**Answer:**
Time series analysis involves studying data points collected or recorded at specific time intervals to identify patterns, trends, and seasonal variations.

Two key components:
1. **Trend:** Long-term movement in data.
2. **Seasonality:** Regular fluctuations repeating over specific periods.

### Question 7: Describe the difference between seasonality and cyclic behavior in time series.
**Answer:**
Seasonality refers to regular, predictable changes that repeat at fixed intervals (e.g., monthly or yearly).
Cyclic behavior represents long-term fluctuations that are not of fixed duration and depend on external factors such as economic cycles.

```python
# Question 8: Write Python code to perform K-means clustering on a sample dataset.
# Answer:
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Create a sample dataset
X, y = make_blobs(n_samples=200, centers=3, random_state=42, cluster_std=1.5)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X')
plt.title('K-Means Clustering Example')
plt.show()
```

```python
# Question 9: What is inheritance in OOP? Provide a simple example in Python.
# Answer:
class Animal:
    def speak(self):
        return 'Some sound'

class Dog(Animal):
    def speak(self):
        return 'Bark'

d = Dog()
print(d.speak())  # Output: Bark
```
